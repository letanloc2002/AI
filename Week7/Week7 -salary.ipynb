```python
import pandas as pd
url='Salary_Data.csv'
data=pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
#mpg:mile per megallon
#disp: thể tích buồng đốt
#hp: horse power 
#wt:weight
```

```python
correlation=data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
sns.pairplot(data)
```

```python
data.plot(kind='scatter',x='YearsExperience',y='Salary',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['YearsExperience']),y=data['Salary'])
print(model.intercept_)
print(model.coef_)
#tìm ra hệ số a và b trong y=ax+b
model.score(X=pd.DataFrame(data['YearsExperience']),y=data['Salary'])
#độ chính xác
```

```python
data.plot(kind='scatter',x='YearsExperience',y='Salary',figsize=(9,7),color='black',xlim=(0,7))
y_pred=model.predict(X=pd.DataFrame(data['YearsExperience']))
plt.plot(data['YearsExperience'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()

model.fit(X=pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2,data['YearsExperience']**3,data['YearsExperience']**4,data['YearsExperience']**5,data['YearsExperience']**6,data['YearsExperience']**7,data['YearsExperience']**8,data['YearsExperience']**9,data['YearsExperience']**10]).T,y=data['Salary'])
print('Intercept of model:',model.intercept_)
print('Coefient of model:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X=pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2,data['YearsExperience']**3,data['YearsExperience']**4,data['YearsExperience']**5,data['YearsExperience']**6,data['YearsExperience']**7,data['YearsExperience']**8,data['YearsExperience']**9,data['YearsExperience']**10]).T,y=data['Salary']))
```

```python
import numpy as np
range=np.arange(1.5,5.5,0.1)

predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X=predictor)
data.plot(kind='scatter',x='YearsExperience',y='Salary',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
plt.plot(data.YearsExperience,data.Salary,color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
model = LinearRegression()
model.fit(X = pd.DataFrame(data['YearsExperience']), y = data['Salary'])
y_pred = model.predict(X = pd.DataFrame(data['YearsExperience'])).round(1)
calculation = pd.DataFrame(np.c_[data.Salary, y_pred], columns = ['Original Salary','Predicted Salary'])
print(calculation)
```
