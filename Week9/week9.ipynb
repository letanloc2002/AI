```python
from sklearn.datasets import fetch_openml
import pandas as pd
import seaborn as sns

# Load the Boston Housing dataset
boston = fetch_openml(data_id=531)

# Convert the dataset to a pandas DataFrame
df = pd.DataFrame(boston.data, columns=boston.feature_names)

# Add the target variable to the DataFrame
df['MEDV'] = boston.target

# Create a pairplot
sns.pairplot(df)
```

```python
from keras.datasets import boston_housing
(x_train,y_train),(x_test,y_test)=boston_housing.load_data()
data=boston_housing.load_data()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
print(x_train_std[:5])
```

```python
model=Sequential()
model.add(Dense(50,activation='relu',input_shape=(13,)))
model.add(Dense(10,activation='linear'))
model.summary()
```

```python
model.compile(optimizer='Adam',loss='mean_absolute_error',metrics=['accuracy'])
model.fit(x_train_std,y_train,epochs=20,batch_size=128)
test_loss,test_acc=model.evaluate(x_test_std,y_test)
print('Do chinh xac tren tap kiem tra: ',test_acc)
print('ham mat mat tren tap kiem tra: ',test_loss)
```
