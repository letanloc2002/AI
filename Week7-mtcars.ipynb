```python
import pandas as pd
url='mtcars.csv'
data=pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
#mpg:mile per megallon
#disp: thể tích buồng đốt
#hp: horse power 
#wt:weight
```

```python
correlation=data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
#xác định giá trị tương quan: số càng lớn thì mức độ liên quan càng lớn, dấu thể hiện thuận nghịch
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python

from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
print(model.intercept_)
print(model.coef_)
#tìm ra hệ số a và b trong y=ax+b
model.score(X=pd.DataFrame(data['wt']),y=data['mpg'])
#độ chính xác
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
y_pred=model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')

```

```python
model=linear_model.LinearRegression()

model.fit(X=pd.DataFrame([data['wt'],data['wt']**2]).T,y=data['mpg'])
print('Intercept of model:',model.intercept_)
print('Coefient of model:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X=pd.DataFrame([data['wt'],data['wt']**2]).T,y=data['mpg']))
```

```python
import numpy as np
range=np.arange(1.5,5.5,0.1)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
predictor=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
