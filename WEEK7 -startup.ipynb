```python
import pandas as pd
url='50_Startups.csv'
data=pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation=data.corr()
print(correlation)
```

```python
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2]).T,y=data['Profit'])
print(model.intercept_)
print(model.coef_)
#tìm ra hệ số a và b trong y=ax+b
model.score(X=pd.DataFrame([data['R&D Spend'],data['R&D Spend']**2]).T,y=data['Profit'])
#độ chính xác
```

```python
import numpy as np
range=np.arange(15,155000,0.1)

predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=predictor)
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black',xlim=(-100,150000))
plt.plot(range,y_pred,color='blue')
```
