```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
```

```python
data_train = pd.read_csv("/content/weather-main.csv")
```

```python
data_train = np.array(data_train)
```

```python
X = data_train[:, :5]
Y_today = data_train[:, 5].astype(int)
# Y_tomor = data_train[:, 6].astype(int)
```

```python
neigh_today = KNeighborsClassifier(n_neighbors=3)
# neigh_tomor = KNeighborsClassifier(n_neighbors=3)
neigh_today.fit(X, Y_today)
# neigh_tomor.fit(X, Y_tomor)
```

```python
temp = 30
wind = 7
cloud = 14
hud = 71
pressure = 1000

if((neigh_today.predict([[temp, wind, cloud, hud, pressure]])[0]) == 1):
  print("Rain Tomorrow: No")
else:
  print("Rain Tomorrow: Yes")

# if((neigh_tomor.predict([[22, 45, 500, 519]])[0]) == 1):
#   print("Rain Tomorrow: No")
# else:
#   print("Rain Tomorrow: Yes")
```
